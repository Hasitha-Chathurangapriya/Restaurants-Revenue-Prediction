```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Restaurant_revenue.csv")
data
```

```python
model = LinearRegression()
model.fit(data[['Number_of_Customers','Menu_Price','Marketing_Spend','Average_Customer_Spending','Promotions','Reviews']],data.Monthly_Revenue)
```

```python
model.predict([[81,34.722067,17.989104,15.482112,1,72]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_new = model.coef_[0]*81 + model.coef_[1]*34.722067 + model.coef_[2]*17.989104 + model.coef_[3]*15.482112+model.coef_[4]*1+model.coef_[5]*72+model.intercept_
y_new
```
